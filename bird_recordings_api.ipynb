```python
import pandas as pd
import numpy as np
import folium
import streamlit as st

import requests
import os
```

```python
puerto_rico_bird_recordings = pd.read_json('puerto_rico_recordings.json')
recordings_df = pd.DataFrame(puerto_rico_bird_recordings['recordings'].tolist())
recordings_df = recordings_df.query('file != "" and lng.notnull() and lat.notnull()')



```

```python

#recordings_df['lat'] = recordings_df['lat'].replace('', np.nan)
recordings_df['lat'] = recordings_df['lat'].astype(float)

#recordings_df['lng'] = recordings_df['lng'].replace('', np.nan)
recordings_df['lng'] = recordings_df['lng'].astype(float)
```

```python

 
    
map_osm = folium.Map(location=[18.22, -66.59], zoom_start=10)

recordings_df.apply(lambda row:folium.Marker(location=[row["lat"], row["lng"]], 
                                              radius=10, interactive = True, 
                                              tooltip=(row['loc']),
                                              popup=f"Lattitude:{row['lat']}<br>"
                                                f"Longitude:{row['lng']}<br>"
                                                 f"Name:{row['loc']}<br>"
                                                 f"Organism:{row['en']}"
                                              ).add_to(map_osm), axis=1)
map_osm
```

```python
url_list = list(recordings_df['file'])
id_list = list(recordings_df['id'])
```

```python

# Function to download file from URL
def download_file(url, save_path):
    response = requests.get(url)
    if response.status_code == 200:
        with open(save_path, 'wb') as file:
            file.write(response.content)
        print(f"Downloaded {save_path}")
    else:
        print(f"Failed to download {url}")

# Function to download files from a list of URLs
def download_files_from_list(url_list, id_list, download_dir):
    os.makedirs(download_dir, exist_ok=True)  # Create download directory if not exist
    for url, file_id in zip(url_list, id_list):
        filename = f"{file_id}.mp3"  # Using id_list for filename
        save_path = os.path.join(download_dir, filename)
        download_file(url, save_path)



# Call the function to download files
download_files_from_list(url_list, id_list, 'audio_files')
```
